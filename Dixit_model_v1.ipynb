# Importing Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

# Loading the Dataset

```python
DataPath = r'C:\Users\dixit\DURHAM\ArtificialIntelligence\Semester 2\AIDI 2004\Week 4\breast+cancer+wisconsin+diagnostic\wdbc.data'
```

```python
data = pd.read_csv(DataPath, header=None) 
```

# Data Preprocessing

```python
# Assign meaningful column names
column_names = [
    'ID', 'Diagnosis', 'Radius1', 'Texture1', 'Perimeter1', 'Area1', 'Smoothness1','Compactness1', 
    'Concavity1', 'Concave_Points1', 'Symmetry1', 'Fractal_Dimension1',
    'Radius2', 'Texture2', 'Perimeter2', 'Area2', 'Smoothness2', 'Compactness2', 
    'Concavity2', 'Concave_Points2', 'Symmetry2', 'Fractal_Dimension2', 
    'Radius3', 'Texture3', 'Perimeter3','Area3', 'Smoothness3', 'Compactness3', 'Concavity3', 
    'Concave_Points3', 'Symmetry3', 'Fractal3_Dimension'
    ]
data.columns = column_names
```

```python
data['Diagnosis'] = data['Diagnosis'].map({'M': 1, 'B': 0})
```

```python
data.drop(columns=['ID'], inplace=True)
```

```python
data.head()
```

# Model Training (Creating the Test Train Split & Scaling the Data)

```python
#Create x and y variables
x = data.drop('Diagnosis', axis=1)
y = data['Diagnosis']

#Create Train and Test Dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20,stratify=y,random_state=50)

#Scale the Data (Standardizing)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train2 = sc.fit_transform(x_train)
x_test2 = sc.transform(x_test)
```

```python
for name,method in [('Logistic Regression', LogisticRegression())]: 
    method.fit(x_train2,y_train)
    predict = method.predict(x_test2)
    print('\nEstimator: {}'.format(name))
    print('\nConfusion Matrix')
    print(confusion_matrix(y_test,predict))
    print('\nClassification Report')
    print(classification_report(y_test,predict))          
```
